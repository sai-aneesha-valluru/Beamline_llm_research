## Important: Keeping Both Repos in Sync

This project uses **two remotes**:
- `origin` → your fork: `sai-aneesha-valluru/beamline_llm` (branch: `Aneesha`) — connected to the Streamlit app
- `upstream` → main repo: `sbhunia/beamline_llm` (branch: `Aneesha`) 

Any code changes must be pushed to **both** so the Streamlit app and the main repo stay in sync.

### One-time setup: Add your fork as a remote

Run this once after cloning (from inside the `beamline_llm` folder, not COLAB):

````python
# Run from beamline_llm/ (one level above COLAB)
!git remote add origin https://github.com/sai-aneesha-valluru/beamline_llm.git
!git remote -v  # verify both remotes are listed
````

After running, you should see four lines — two for `origin` (your fork) and two for `upstream` (main repo):

```
origin    https://github.com/sai-aneesha-valluru/beamline_llm.git (fetch)
origin    https://github.com/sai-aneesha-valluru/beamline_llm.git (push)
upstream  https://github.com/sbhunia/beamline_llm.git (fetch)
upstream  https://github.com/sbhunia/beamline_llm.git (push)
```

### Every time you make changes: Push to both remotes

After editing any files, run the following in your terminal:

````python
!git add .
!git commit -m "describe your change here"
!git push origin Aneesha      # pushes to your fork → updates Streamlit app
!git push upstream Aneesha    # pushes to main repo → keeps sbhunia/beamline_llm in sync
````

> Always push to **both** remotes. If you only push to `origin`, the main repo falls behind. If you only push to `upstream`, the Streamlit app won't pick up the changes.

> The Streamlit app auto-redeploys from the `Aneesha` branch of `origin` (your fork) whenever changes are pushed to it.
